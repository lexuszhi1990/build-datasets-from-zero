```python
import csv
import numpy as np

import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib.colors as colors

import cv2

import os
import requests
```

```python
root_dir='/mnt/openimgv4/'
lbl_dir=root_dir+'valid/validation-annotations-bbox.csv'
cls_dir_num=root_dir+'clsDict_new.csv'
cls_dir_code=root_dir+'train/label/challenge-2018-class-descriptions-500.csv'
dat_dir=root_dir+'sources/train_img'
download_dir=root_dir+'downloads/'
```

```python
# loop through all files in dat_dir (jpg)
list = os.listdir(dat_dir)
img_idx = 0
n=1
for i in range(0,len(list)):
    path = os.path.join(dat_dir,list[i])
    if os.path.isfile(path) and path[-5:]=='gstmp':
        file_id=list[i][:-11]
        print('downloading:',n,file_id)
        image_url = "https://requestor-proxy.figure-eight.com/figure_eight_datasets/open-images/train/"+str(file_id)+".jpg"
        print(image_url)
        r = requests.get(image_url) # create HTTP response object
        with open(download_dir+str(file_id)+".jpg",'wb') as f:
            f.write(r.content)
        n+=1
            
```
