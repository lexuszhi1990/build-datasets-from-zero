```python
import csv
import numpy as np

import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib.colors as colors

import cv2

import os
```

```python
target_labels=[113, 164, 360, 418, 338] #merger
```

```python
root_dir='/mnt/openimgv4/'
lbl_dir=root_dir+'valid/validation-annotations-bbox.csv'
cls_dir_num=root_dir+'clsDict_new.csv'
cls_dir_code=root_dir+'train/label/challenge-2018-class-descriptions-500.csv'
dat_dir=root_dir+'valid/val_val_data/'
```

```python
# convert csv file to dict
def row_csv2dict(csv_file):
    dict_club={}
    with open(csv_file)as f:
        reader=csv.reader(f,delimiter=',')
        for row in reader:
            dict_club[row[0]]=row[1]
    return dict_club
```

```python
# search for class name for a given label
def get_cls(lbl):
    cls_dict=row_csv2dict(cls_dir_code)
    if lbl in cls_dict:
        return cls_dict[lbl]
    else:
        return 'None' # return 'None' for empty labels
```

```python
# convert code to actual class name
# (function from visual)
def code2name(labels):
    for i in range(np.size(labels,0)):
        labels[i,0]=get_cls(labels[i,0])
    return labels
```

```python
# convert class name to number index
# (simplifying the labels)
def name2num(labels):
    cls_dict_num=row_csv2dict(cls_dir_num)
    #
    for i in range(np.size(labels,0)):
        if labels[i,0] in cls_dict_num:
            labels[i,0]=cls_dict_num[labels[i,0]]
        else:
            labels[i,0]=-1 # invalid bbox
    return labels
```

```python
# search for labels for a given image id
id_idx=0 #index of image id
lbl_idx=[2, 4, 5, 6, 7] #label_name, xmin, xmax, ymin, ymax

# only preserve non-group labels
gp_idx=10

def get_box(id,csv_dat):
    out=np.reshape(['0',0,0,0,0],(1,5))
    for i,rows in enumerate(csv_dat):
        if rows[id_idx]==id and int(rows[gp_idx])==0: # only preserve non-group labels
            tmp=['0']
            row=rows
            for k in lbl_idx:
                tmp.append(str(row[k]))
            tmp=tmp[1:]
            out=np.append(out,np.reshape(tmp,(1,5)),axis=0)
    out=np.delete(out,0,axis=0)
    return out # every 5 values (1 row) corresponds to 1 bbox
```

```python
# loop through all files in dat_dir (jpg)
list = os.listdir(dat_dir)
img_idx = 0
with open(lbl_dir,encoding='utf-8') as cfile:
    reader = csv.reader(cfile)
    readeritem=[]
    readeritem.extend([row for row in reader]) 
with open(root_dir+'valid/val_val_dat_pp_merge_ng.lst', 'w+') as f:
    for i in range(0,len(list)):
        if i%50==0:
            print(i)
        path = os.path.join(dat_dir,list[i])
        if os.path.isfile(path) and path[-3:]=='jpg':
            file_id=list[i][:-4]
            lbls=name2num(code2name(get_box(file_id,readeritem)))
            #print(get_box(file_id,csv_dat))
            img=cv2.imread(path)
            W=np.size(img,1)
            H=np.size(img,0)
            # generate the multi-box string
            mcstr=''
            box_count=0
            for j in range (np.size(lbls,0)):
                if int(lbls[j,0]) in target_labels: # '-1' is invalid label
                    mcstr=mcstr+str(0)+'\t' #target_labels.index(int(lbls[j,0]))
                    mcstr+=str(lbls[j,1]) + '\t' + str(lbls[j,3]) + '\t' 
                    mcstr+=str(lbls[j,2]) + '\t' + str(lbls[j,4]) + '\t'
                    box_count+=1
            if box_count>0:
                f.write(
                str(img_idx) + '\t' +
                str(4) + '\t' + str(5) + '\t' +
                str(W) + '\t' + str(H) + '\t' +
                mcstr +
                str(file_id) + '.jpg\n'
                )
                img_idx+=1
```
