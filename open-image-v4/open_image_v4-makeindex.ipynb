```python
import csv
```

```python
root_dir='/mnt/openimgv4/'
cls_dir=root_dir+'train/label/challenge-2018-class-descriptions-500.csv'
```

```python
# convert code to number index
def row_csv2csv(csv_file,isPrint):
    with open(csv_file)as f:
        reader=csv.reader(f,delimiter=',')
        wf=open(root_dir+'clsDict_new.csv','w', newline='')
        writer=csv.writer(wf)
        i=0
        for row in reader:
            writer.writerow([row[1],i])
            if isPrint:
                print([row[1],i])
            i+=1
    print ('job\'s done!')
```

```python
row_csv2csv(cls_dir,1)
```
